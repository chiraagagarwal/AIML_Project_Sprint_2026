```python
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({'Missing_Count': missing, 'Percentage': missing_pct})
print(missing_df[missing_df['Missing_Count'] > 0])
```

```python
duplicates = df.duplicated().sum()
print(f"Exact duplicate rows: {duplicates}")
```

```python
name_dupes = df['name'].duplicated().sum()
print(f"Duplicate phone names: {name_dupes}")
```

```python
if name_dupes > 0:
    print(df[df['name'].duplicated(keep=False)].sort_values('name'))
```

```python
df.columns = df.columns.str.lower()
print(df.columns.tolist())
```

```python
df['price_clean'] = df['price'].str.replace('₹', '').str.replace(',', '').astype(float)
print(df[['price', 'price_clean']].head())
print(f"Price range: {df['price_clean'].min()} - {df['price_clean'].max()}")
```

```python
df['battery_mah'] = df['battery'].str.extract(r'(\d+)\s*mAh').astype(float)
print(df[['battery', 'battery_mah']].head())
```

```python
df['ram_gb'] = df['storage'].str.extract(r'(\d+)\s*GB\s*RAM').astype(float)

df['storage_gb'] = df['storage'].str.extract(r'(\d+)\s*GB\s*inbuilt').astype(float)

print(df[['storage', 'ram_gb', 'storage_gb']].head())
```

```python
df['display_size'] = df['display_info'].str.extract(r'(\d+\.?\d*)\s*inches').astype(float)

df['refresh_rate'] = df['display_info'].str.extract(r'(\d+)\s*Hz').astype(float)

print(df[['display_info', 'display_size', 'refresh_rate']].head())
```

```python
def clean_memorycard(val):
    if pd.isnull(val):
        return 'Not_Specified'
    val = str(val).lower()
    if 'not' in val and 'supported' in val:
        return 'Not_Supported'
    elif 'hybrid' in val:
        return 'Hybrid'
    elif 'supported' in val or 'support' in val:
        return 'Supported'
    else:
        return 'Not_Specified'

df['memorycard_clean'] = df['memorycard_support'].apply(clean_memorycard)
print(df['memorycard_clean'].value_counts())
```

```python
print(df[['spec_score', 'price_clean', 'battery_mah', 'ram_gb', 'storage_gb', 'display_size', 'refresh_rate']].dtypes)
print(f"\nNumeric columns preview:\n{df[['spec_score', 'price_clean', 'battery_mah', 'ram_gb']].describe()}")
```

```python
df_cleaned = df.drop_duplicates(subset=['name'], keep='first')
print(f"Rows after removing duplicates: {len(df_cleaned)}")
```

```python
df_cleaned['battery_mah'].fillna(df_cleaned['battery_mah'].median(), inplace=True)
df_cleaned['ram_gb'].fillna(df_cleaned['ram_gb'].median(), inplace=True)
df_cleaned['storage_gb'].fillna(df_cleaned['storage_gb'].median(), inplace=True)

print(df_cleaned[['battery_mah', 'ram_gb', 'storage_gb']].isnull().sum())
```

```python
# Select relevant columns and export
columns_to_keep = ['name', 'price_clean', 'spec_score', 'battery_mah', 'ram_gb', 
                   'storage_gb', 'display_size', 'refresh_rate', 'memorycard_clean', 'os', 'sim_info']

df_final = df_cleaned[columns_to_keep].copy()
df_final.columns = ['phone_name', 'price', 'spec_score', 'battery_mah', 'ram_gb', 
                    'storage_gb', 'display_size_inches', 'refresh_rate_hz', 'memory_card_support', 'os', 'sim_info']

df_final.to_csv('cleaned_dataset.csv', index=False)
print(f"Cleaned dataset exported: cleaned_dataset.csv ({len(df_final)} rows)")
```
